En el mundo actual la unica manera de mantener el mundo intercomunicado con las tasas de transmicion requeriadas a nivel intercontinental es por medio de red MPLS las cuales son el backbone de la internet como la conocemos hoy, en esta tarea tendremos que simular diferentes redes MPLS, con el proposito de conocer como se comporta cada una de ellas frenete a diferentes condiciones.


    
## El modelo

Los usuarios de la red MLPS seran reprecentados por medio de un modelo ON-OFF, esto significa
que mientras se encuentra el usuario en modo ON este se encuentra transmitiendo a su
m치xima tasa de bits mientras que en modo OFF el usuario no transmite.

Ademas la red se compone de un conjunto de nodos interconectados entre si por medio de enlases, antes de comenzar la coneccion sera nesesario de que todos los enlases entre los 2 nodos involucrados se encuentren disponibles, de lo contrario se considera el enlase blockeado. 

## Supuestos 

Se considera a los usuarios como individuos, impidiedo que estos accedan al servivio mientras aun no aun sido terminados de atender.


La simulaci칩n se efectu칩 con la ayuda del lenguaje de programaci칩n C++.

## Compilacion 
	Ejecutar el compilador de el sistema, se recomienda g++
	usango g++: 
		$ g++ main.cpp

## Ejecucion 
	en linux:
		$ ./a.out {Nombre de topologuia (string)} {Capacidad canal (int)} {criterio de parada (int)} {Ton (float)} {Carga promedio (float)}

Todas las simulaciones a continuacion tienen como condicion de parada 1000000 llegadas, ademas de tener un Ton de 0.001 y una carga promedio de 0.3.





```python
#Programador: Mario Andres Marin Soto, ROL:201321048-1
#Asignatura: TEL341
#Fecha de entrega: yes

import sys
from subprocess import call,Popen, PIPE, STDOUT
import matplotlib.pyplot as plt
import numpy as np
```

```python
call("g++ main.cpp",shell=True)
```

```python
file_list = ["NSFNet","ArpaNet","EON","EuroLarge","EuroCore","UKNet"]
capacity_start = [13,33,18,89,4,21]
capacity_end = [7,15,9,30,3,11]

simulation_data = [[],[],[],[],[],[]] #arreglo con todos los datos de las simulaciones

for x in range(0,6): #ejecucion de las simulaciones y almacenamiento de los datos
    for i in range(capacity_start[x],capacity_end[x]-1,-1):
        command = "./a.out "+file_list[x]+" "+str(i)+" 1000000 0.001 0.3"
        p = Popen(command, stdin=PIPE, stdout=PIPE, stderr=STDOUT,shell=True)
        output=p.stdout.read()
        text = ''.join(chr(i) for i in output)
        lines = text.splitlines()
        data = [float(i) for i in lines]
        simulation_data[x].append(data)
        
prob_list = [[],[],[],[],[],[]] #arreglo que almacena datos para los graficos
counter = 0;
for x in simulation_data: #se extrae la prob de bloqueo de las redes (el ultimo dato)
    for i in x:
        prob_list[counter].append(i[-1])
        del i[-1]
    counter = counter + 1
counter = 0;
for x in prob_list: #se grafica los datos previemante extraidos
    plt.figure(counter)
    plt.grid(color="black")
    plt.title("Simulation of "+file_list[counter]+ " network",fontsize=18)
    plt.xlabel('Capacity of the links')
    plt.ylabel('Blocking probability')
    t=list(range(capacity_start[counter],capacity_end[counter]-1,-1))
    plt.plot(t,x,label="Simulation")
    plt.subplots_adjust(top=1.5,right=1.5)
    plt.show()
    counter = counter +1
    
last_graph = [[],[],[],[],[],[],[],[]] #arreglo con los datos requeridos para el ultimo grafico

for x in simulation_data: #se procede a extraer los datos relevantes para el ultimo grafico
    # para los puntos del grafico donde no existen datos por la estructura de la topologuia
    # se rellena con 0. No es elegante pero es funcional.
    if len(x[-1]) == 1:
        last_graph[0].append(x[-1][0])
        last_graph[1].append(0)
        last_graph[2].append(0)
        last_graph[3].append(0)
        last_graph[4].append(0)
        last_graph[5].append(0)
        last_graph[6].append(0)
        last_graph[7].append(0)
    elif len(x[-1]) == 2:
        last_graph[0].append(x[-1][0])
        last_graph[1].append(x[-1][1])
        last_graph[2].append(0)
        last_graph[3].append(0)
        last_graph[4].append(0)
        last_graph[5].append(0)
        last_graph[6].append(0)
        last_graph[7].append(0)
    elif len(x[-1]) == 3:
        last_graph[0].append(x[-1][0])
        last_graph[1].append(x[-1][1])
        last_graph[2].append(x[-1][2])
        last_graph[3].append(0)
        last_graph[4].append(0)
        last_graph[5].append(0)
        last_graph[6].append(0)
        last_graph[7].append(0)
    elif len(x[-1]) == 4:
        last_graph[0].append(x[-1][0])
        last_graph[1].append(x[-1][1])
        last_graph[2].append(x[-1][2])
        last_graph[3].append(x[-1][3])
        last_graph[4].append(0)
        last_graph[5].append(0)
        last_graph[6].append(0)
        last_graph[7].append(0)
    elif len(x[-1]) == 5:
        last_graph[0].append(x[-1][0])
        last_graph[1].append(x[-1][1])
        last_graph[2].append(x[-1][2])
        last_graph[3].append(x[-1][3])
        last_graph[4].append(x[-1][4])
        last_graph[5].append(0)
        last_graph[6].append(0)
        last_graph[7].append(0)
    elif len(x[-1]) == 6:
        last_graph[0].append(x[-1][0])
        last_graph[1].append(x[-1][1])
        last_graph[2].append(x[-1][2])
        last_graph[3].append(x[-1][3])
        last_graph[4].append(x[-1][4])
        last_graph[5].append(x[-1][5])
        last_graph[6].append(0)
        last_graph[7].append(0)
    elif len(x[-1]) == 7:
        last_graph[0].append(x[-1][0])
        last_graph[1].append(x[-1][1])
        last_graph[2].append(x[-1][2])
        last_graph[3].append(x[-1][3])
        last_graph[4].append(x[-1][4])
        last_graph[5].append(x[-1][5])
        last_graph[6].append(x[-1][6])
        last_graph[7].append(0)
    elif len(x[-1]) == 8:
        last_graph[0].append(x[-1][0])
        last_graph[1].append(x[-1][1])
        last_graph[2].append(x[-1][2])
        last_graph[3].append(x[-1][3])
        last_graph[4].append(x[-1][4])
        last_graph[5].append(x[-1][5])
        last_graph[6].append(x[-1][6])
        last_graph[7].append(x[-1][7])

plt.figure(100)
plt.grid(color="black")
plt.title("Final simulation",fontsize=18)
plt.xlabel('Networks')
plt.ylabel('Blocking probability')
plt.plot(file_list,last_graph[0],label="1 hops")
plt.plot(file_list,last_graph[1],label="2 hops")
plt.plot(file_list,last_graph[2],label="3 hops")
plt.plot(file_list,last_graph[3],label="4 hops")
plt.plot(file_list,last_graph[4],label="5 hops")
plt.plot(file_list,last_graph[5],label="6 hops")
plt.plot(file_list,last_graph[6],label="7 hops")
plt.plot(file_list,last_graph[7],label="8 hops")
plt.legend(loc = 'best')
plt.subplots_adjust(top=1.5,right=1.5)
plt.show()
```
